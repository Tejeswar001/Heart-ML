### Separate Featires Types

```python
categorical_cols = [i for i in df.columns if df[i].dtype == 'object']
numerical_cols = [i for i in df.columns if df[i].dtype != 'object']
```

```python
categorical_cols
```

```python
numerical_cols
```

### Encode Categorical Features

```python
df[categorical_cols]
```

```python
SEX_MAP = {'F':0, 'M':1}
YES_NO = {'N':0, 'Y':1}
RISK_MAP = {'LOW':0, 'INTERMEDIARY':1, 'HIGH':2}
BP_MAP = {
    'Normal':0,
    'Elevated':1,
    'Hypertension Stage 1':2,
    'Hypertension Stage 2':3
}

df['Sex'] = df['Sex'].map(SEX_MAP)
df['Smoking Status'] = df['Smoking Status'].map(YES_NO)
df['Diabetes Status'] = df['Diabetes Status'].map(YES_NO)
df['Family History of CVD'] = df['Family History of CVD'].map(YES_NO)
df['Physical Activity Level'] = df['Physical Activity Level'].map({'Low':0,'Moderate':1,'High':2})
df['CVD Risk Level'] = df['CVD Risk Level'].map(RISK_MAP)
df['Blood Pressure Category'] = df['Blood Pressure Category'].map(BP_MAP)
```

```python
df.info()
```

```python
df.head()
```

### Feature Scaling

Why?

- Needed for Logistic Regression, SVM, KNN

- Required for Clustering

- Helps Regression stability

Let's use StandardScaler.

```python
scaler = StandardScaler()
```

```python
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```
